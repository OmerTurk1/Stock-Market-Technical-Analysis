```python
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
hist.to_csv("hisse_verisi.csv")
```

```python
zaman_adimi = 10        # Gemi ka gne bakarak tahmin yaplacak?
ozellik_sayisi = 1      # Ka farkl deiken kullanlyor? (imdilik sadece "close" kullanld iin 1)
test_orani = 0.2        # Verinin ne kadar test iin ayrlacak?
epoch_sayisi = 50       # Eitim tur says
batch_boyutu = 16       # Her admda ilenecek veri grubu bykl
```

```python
data = hist['Close'].values.reshape(-1, 1) # LSTM 2D veya 3D veri ister
scaler = MinMaxScaler(feature_range=(0, 1)) # Deerler 0 ve 1 arasnda olacak
scaled_data = scaler.fit_transform(data)
scaled_data
```

```python
def create_dataset(dataset, time_step):
    X, y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step), 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

X, y = create_dataset(scaled_data, zaman_adimi)

# rnek Says otomatik olarak veriden hesaplanr
ornek_sayisi = X.shape[0]
print("rnek says:",ornek_sayisi)
X = X.reshape(ornek_sayisi, zaman_adimi, ozellik_sayisi)
```

```python
split = int(ornek_sayisi * (1 - test_orani))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
print(f"Veri Yaps: [rnek: {ornek_sayisi}, Zaman Adm: {zaman_adimi}, zellik: {ozellik_sayisi}]")

model = Sequential([
    # lk LSTM katman
    LSTM(units=50, return_sequences=True, input_shape=(zaman_adimi, ozellik_sayisi)),
    Dropout(0.2), # overfitting engellemek iin

    # kinci LSTM katman
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),

    # k katman (Saysal bir deer dnecei iin tek bir nron)
    Dense(units=1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=epoch_sayisi,
    batch_size=batch_boyutu,
    verbose=1
)
```

```python
tahminler = model.predict(X_test)
# leklendirmeyi geri al (0-1 arasndan gerek fiyatlara dn)
tahminler = scaler.inverse_transform(tahminler)
gercek_degerler = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
plt.figure(figsize=(8,6))
plt.plot(gercek_degerler,color="blue",label="Gerek Deerler")
plt.plot(tahminler,color="orange",label="Tahmin Deerler")
plt.legend()
```
